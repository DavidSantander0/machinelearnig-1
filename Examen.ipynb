## 01 - 1 pt
- Dado un array de elementos, escribir una función, que reciba dicho array como parámetro de entrada y devuelva el elemento repetido. Usando plain Python, sin ninguna librería.

```python
array_src = [4, 8, 0, 3, 6, 2, 1, 5, 3, 9, 7]   
rep_element = repeated_element(array_src)
print(rep_element)# This calls your function repeated_element# This will fail if the element is not found correctly
```

```python
# Your solution HERE !!!!
def repeated_element(arr):
    seen=set()
    result=[]
    for num in arr:
        if num in seen:
            result.append((num,True))
        else:
            seen.add(num)
            
    return result

    
```

- Crear una lista de meses de Enero a Diciembre (eje X)
- Generar datos aleatorios para el eje Y, hacerlo 3 veces distintas (Y0, Y1 e Y2)
- Representar las 3 secuencias de datos aleatorios en una sola figura usando matplotlib
- Añadir una leyenda para poder identificar cada secuencia
- Nombrar las 3 secuencias de la siguiente manera ("Facebook", "Twitter", "Instagram")
- Añadir un título y nombres a los dos ejes
- Añadir una anotación (texto y flecha) a uno de los gráficos

```python

# Semilla opcional para reproducibilidad
np.random.seed(42)

# Crear tres listas de datos Y aleatorios
Y0 = np.random.randint(0, 100, size=10)
Y1 = np.random.randint(0, 100, size=10)
Y2 = np.random.randint(0, 100, size=10)

# Crear eje X común para las tres secuencias
X = np.arange(10)

#ESTO CONSULTADO CON CHATI 
plt.figure(figsize=(10, 6))  # Tamaño de la figura

plt.plot(X, Y0, label="Facebook", marker='o')
plt.plot(X, Y1, label="Twitter", marker='s')
plt.plot(X, Y2, label="Instagram", marker='^')

plt.legend()
plt.title("Actividad de redes sociales 📊")
plt.xlabel("Tiempo (días)")
plt.ylabel("Interacciones")
# Añadir anotación en el punto más alto de Y2 (Instagram)
max_idx = np.argmax(Y2)
plt.annotate("Máximo de Instagram",
             xy=(X[max_idx], Y2[max_idx]),
             xytext=(X[max_idx]+1, Y2[max_idx]+20),
             arrowprops=dict(arrowstyle="->", color='purple'),
             fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
