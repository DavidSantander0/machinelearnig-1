# Resumen: Tema 3 - Extracción de Características



## 1. Estadísticos Unidimensionales

Esta sección fundamenta las bases del análisis estadístico sobre datos de una sola dimensión.

### 1.1 Conceptos Básicos
* **Población:** Es el conjunto total de ítems (instancias, individuos, elementos o filas) sobre el cual se desea estudiar o extraer conclusiones (ej. todos los lotes de un producto industrial).
* **Muestra:** Es un subconjunto representativo de la población. Se obtiene mediante un muestreo aleatorio para analizar la población sin necesidad de evaluarla por completo, ahorrando tiempo y costes.
* **Inferencia Estadística:** Es el proceso lógico y matemático por el cual, analizando la muestra, se obtienen conclusiones sobre la población total con un margen de error métrico conocido.

### 1.2 Clasificación de Variables
Una variable estadística es cada cualidad, propiedad o atributo que poseen los individuos de la población y que puede cambiar. Se dividen en dos grandes familias:

1. **Cualitativas (Categóricas):** No son numéricas.
    * *Ordinales:* Existe un criterio de orden implícito y lógico (ej. grado de satisfacción del cliente, puesto en una competición).
    * *Nominales:* Modalidades puras sin ningún tipo de criterio de orden (ej. sexo, estado civil).

2. **Cuantitativas (Numéricas):** Atributos puramente numéricos.
    * *Discretas:* Toman un número finito y contable de valores enteros (ej. cantidad de hijos, número de empleados).
    * *Continuas:* Asumen un número infinito de valores dentro de un rango continuo (ej. estatura, peso corporal, temperatura).

### 1.3 Análisis de Variables Cualitativas y Cuantitativas

**Análisis de Variables Cualitativas:**
* **Tablas de Frecuencia:** Organizan los datos mostrando la frecuencia absoluta (conteo directo), frecuencia relativa (proporción sobre 1), absoluta acumulada y relativa acumulada.
* **Gráficos:** Diagramas de barras (para comparar frecuencias absolutas) y Diagramas de sectores/pastel (para visualizar porcentajes).

**Análisis de Variables Cuantitativas:**
* **Medidas de Tendencia Central:** Media (promedio matemático), Mediana (valor central al ordenar los datos) y Moda (valor que más se repite).
* **Medidas de Posición:** Cuartiles (Q1=25%, Q3=75%). El Rango Intercuartílico (RIC = Q3 - Q1) contiene el 50% central de la muestra. Se usa para crear Boxplots y detectar valores atípicos (outliers) usando umbrales como $RIC \cdot 1.5$.
* **Medidas de Dispersión:** Varianza ($\sigma^2$) y Desviación Típica ($\sigma$). Miden cuánto se alejan los datos individuales de la media general.
* **Medidas de Distribución:**
    * *Coeficiente de Asimetría:* Evalúa si la curva de distribución está sesgada a la derecha (positiva), izquierda (negativa) o si es simétrica.
    * *Curtosis:* Mide el grado de 'achatamiento' o apuntamiento de la distribución frente a una curva normal perfecta (Platicúrtica, Mesocúrtica, Leptocúrtica).

## 2. Características Bidimensionales (Visión Artificial)

### 2.1 Introducción a los Datos 2D
Una imagen digital es interpretada por el ordenador como una función bidimensional $f(x,y)$, donde $x$ e $y$ representan las coordenadas espaciales en una matriz, y el valor de la función dictamina la intensidad (luminosidad o color) del píxel.

### 2.2 Procesado de Imagen
Fase de limpieza previa a la extracción de características.

* **Transformaciones de Intensidad:** Sumar o multiplicar valores de la matriz para lograr cambios de brillo, contraste, calcular el negativo o ecualizar histogramas para balancear la luz.
* **Espacios de Color:** Conversiones algorítmicas entre modelos:
    * *RGB a Grises:* Simplifica la imagen a un canal de intensidad.
    * *CMYK:* Modelo sustractivo para impresión.
    * *HSV:* Separa el Matiz (Hue), Saturación y Valor de luminosidad.
    * *Lab y YCrCb:* Aíslan la luminosidad pura de la crominancia (color), excelentes para evitar que las sombras alteren la detección algorítmica.
* **Transformaciones Geométricas:** Traslación (desplazamiento), Cropping (recorte para aislar la Región de Interés o ROI), Rotación, Perspectiva (corregir deformaciones 3D a 2D) y Flipping (volteos axiales).

### 2.3 Segmentación de Imágenes
Simplifica la imagen asignando etiquetas comunes a píxeles con propiedades similares, aislando los objetos del fondo.

* **Tipos:** *Semántica* (clasifica grupos enteros, ej. todos los coches) y *De Instancia* (detecta individuos únicos, ej. coche 1, coche 2).
* **Técnicas Principales:**
    1. **Thresholding (Umbralización):** Destaca el **Método de Otsu**, un algoritmo que busca automáticamente el umbral matemático que maximiza la varianza entre el objeto principal y el fondo analizando el histograma.
    2. **Detección de Bordes:** Uso del Operador Sobel para calcular gradientes (derivadas matriciales) en X e Y. Posteriormente, el algoritmo de Canny lo refina con supresión de no-máximos y umbralización por histéresis para conectar bordes rotos.
    3. **Otras:** Clustering K-Means y Segmentación semántica mediante Redes Neuronales Convolucionales (Autoencoders).

### 2.4 Descriptores de Forma
Métricas geométricas exactas aplicadas a objetos binarizados:

* **Dimensiones Directas:** Área (píxeles internos) y Perímetro (píxeles de la frontera exterior).
* **Envolventes:** *Bounding Box* (rectángulo geométrico exacto más pequeño que contiene el objeto) y *Convex Hull* (polígono convexo más ajustado posible).
* **Proporciones Estructurales:**
    * *Excentricidad:* Evalúa matricialmente si la forma tiende a un círculo (0) o a una línea (1).
    * *Extensión, Solidez y Compacidad:* Ratios que dividen el área del objeto entre el área de sus envolventes para inferir qué tan robusto, circular o "dentado" es.
    * *Orientación:* Grados precisos de inclinación del eje mayor del objeto.

### 2.5 Descriptores de Textura
Extraen patrones visuales complejos invariantes a condiciones externas que la geometría no puede captar:

1. **Matriz de Coocurrencias:** Algoritmo que evalúa cuántas veces un nivel de gris está adyacente a otro nivel distinto. Extrae métricas de contraste, disimilitud, homogeneidad y energía (ideal para diferenciar texturas como asfalto vs. césped).
2. **SIFT (Scale-Invariant Feature Transforms):** Detecta "puntos clave" inmutables a cambios de escala o rotación. Crea pirámides de Diferencias de Gaussianas (DoG), filtra extremos usando la Matriz Hessiana y genera un fingerprint de **128 características numéricas** basado en histogramas de orientación.
3. **SURF (Speeded-Up Robust Features):** Evolución diseñada para ser hasta 3 veces más veloz que SIFT (ideal para tiempo real). Aproxima los cálculos matemáticos usando "Imágenes Integrales", generando descriptores de 64 o 128 características rápidamente.
4. **HOG (Histogram of Oriented Gradients):** Estándar en la detección de siluetas humanas/peatones. Divide la imagen en pequeñas celdas de 8x8 píxeles, calcula la dirección angular del gradiente (pendiente de color) en cada una, y devuelve un modelo numérico masivo y altamente robusto al ruido.
5. **LBP (Local Binary Patterns):** Desliza una ventana que compara el brillo del píxel central exclusivamente con sus vecinos. Si el vecino es más brillante, anota '1', si es más oscuro, '0'. Esto genera patrones binarios (ej. 8 bits). Destacan los **LBP uniformes** (aquellos con 2 o menos saltos de 0 a 1), los cuales eliminan información redundante y son el estándar de oro para la identificación facial y catalogación de texturas orgánicas.
